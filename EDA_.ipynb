## 수치형 변수 > 범주형 변수

```python
# 방법 1
def get_category(age):
    cat = ''
    if age <= -1: cat = 'Unknown'
    elif age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'

    return cat

df['Age'].fillna(df['Age'].mean(),inplace=True)# NA 때문에 범주화가 진행안됌
age_categorized = df['Age'].apply(lambda x : get_category(x))
age_categorized
```

```python
# 방법 2
bins = [np.min(df.Age), 5, 12, 18, 25, 35, 60, np.max(df.Age)]
labels = ["Baby", "Child", "Teenager", "Student", " Young Adult", "Adult", "Elderly"]

df_cut = pd.cut(df["Age"], bins, labels=labels)
df_cut
```

## 범주형 변수의 encoding

### Label Encoding

```python
from sklearn import preprocessing

def encode_features(dataDF):
    features = ['Cabin', 'Sex', 'Embarked'] # 인코딩할 변수명
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature]) # 인코딩한 값 저장.

    return dataDF

df_label = encode_features(df)
df_label.head()
```

### One-Hot Encoding

```python
import pandas as pd
train = pd.DataFrame({'num1':[1,2,3,4,5], 'num2':[10,20,30,40,50], 'cat1':['a', 'a', 'b', 'c', 'c']})
train
```

```python
pd.get_dummies(train, dtype=int)
```

1. pd.get_dummies()
- pandas.get_dummies는 train 데이터의 특성을 학습하지 않기 때문에 train 데이터에만 있고 test 데이터에는 없는 카테고리를 test 데이터에서 원핫인코딩 된 칼럼으로 바꿔주지 않는다.
- 따라서 train, test가 분리되어 있다면 합쳐준 다음에 작업을해야함.
- 그게 싫으면 sklearn의 OneHotEncoder을 활용

```python
test = pd.DataFrame({'num1':[1,2,3,4,5], 'num2':[10,20,30,40,50], 'cat1':['a', 'a', 'b', 'b', 'a']})
test
```

```python
pd.get_dummies(test, dtype=int)
```

2. sklearn OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

# train data
ohe = OneHotEncoder(sparse_output = False) # sparse 파라미터 바뀜.
# fit_transform은 train에만 사용하고 test에는 학습된 인코더에 fit만 해야한다
train_cat = ohe.fit_transform(train[['cat1']])
train_cat
```

```python
ohe.categories_
```

```python
pd.DataFrame(train_cat, columns=['cat1_' + col for col in ohe.categories_[0]])
```

```python
pd.concat([train.drop(columns=['cat1']),
           pd.DataFrame(train_cat, columns=['cat1_' + col for col in ohe.categories_[0]])], axis=1)
```

```python
# test data : train으로 fit했기 때문에 transform만 사용하면 된다.
test_cat = ohe.transform(test[['cat1']])
test_cat
```

```python
cat1 = pd.DataFrame(test_cat, columns=['cat1_' + col for col in ohe.categories_[0]])
cat1
```

```python
pd.concat([test.drop(columns=['cat1']), cat1], axis=1)
```
